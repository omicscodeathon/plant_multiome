```python
from Bio import Entrez
import pandas as pd
import os
```

```python
def get_tax_id(species):
    species = species.replace(" ", "+").strip()
    search = Entrez.esearch(term = species, db = "taxonomy", retmode = "xml")
    record = Entrez.read(search)
    return record['IdList'][0]
###############################################
def get_tax_data(taxids):
     dd = {'Superkingdom':[],
         'Phylum':[],
         'Class':[],
         'Order':[],
         'Family':[],
         'Genus':[]}
    for id in taxids:
        search = Entrez.efetch(id = taxid, db = "taxonomy", retmode = "xml")
        taxdata = Entrez.read(search)
        for taxa in taxdata[0]['LineageEx']:
            if taxa['Rank'] == 'superkingdom':
                dd['Superkingdom'].append(taxa['ScientificName'])
                
    lineage = {d['Rank']:d['ScientificName'] for d in
        taxdata[0]['LineageEx'] if d['Rank'] in ['superkingdom','phylum','class','order', 'family','genus','species']}
    return lineage
```
