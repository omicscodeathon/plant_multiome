```python
import pandas as pd
import seaborn as sb 
import matplotlib.pyplot as plt
```

```python
animals = pd.read_csv('drive/MyDrive/Codeathon/Selected_animals.csv')
```

```python
animals = animals.loc[animals['Genome.CDS'] > 30]
```

```python
fig = plt.figure(figsize=(100, 30))
ax = plt.subplot(1, 3, 1)
sb.histplot(animals['Phylum'])
plt.show()
```

```python
fig = plt.figure(figsize=(100, 30))
ax = plt.subplot(1, 3, 1)
sb.histplot(animals['Family'])
plt.xticks(rotation=90)
plt.show()
```

```python
fig = plt.figure(figsize=(100, 30))
ax = plt.subplot(1, 3, 1)
sb.boxplot(x="Family", y="size.ratio", data=animals,)
plt.xlabel("Family")
plt.xticks(rotation=90)
plt.ylabel("Genome size ratio (Genome/Mitochondrio)")
ax.set_yscale('log')
plt.show()
```

```python
fig = plt.figure(figsize=(100, 30))
ax = plt.subplot(1, 3, 1)
sb.boxplot(x="Phylum", y="size.ratio", data=animals,)
plt.xlabel("Phylum")
plt.xticks(rotation=90)
plt.ylabel("Genome size ratio (Genome/Mitochondrio)")
ax.set_yscale('log')
plt.show()
```

```python
fig = plt.figure(figsize=(100, 30))
ax = plt.subplot(1, 3, 1)
sb.boxplot(x="Genus", y="size.ratio", data=animals,)
plt.xlabel("Genus")
plt.xticks(rotation=90)
plt.ylabel("Genome size ratio (Genome/Mitochondrio)")
ax.set_yscale('log')
plt.show()
```

```python
fig = plt.figure(figsize=(100, 30))
ax = plt.subplot(1, 3, 1)
sb.boxplot(x="Phylum", y="Genome.GC", data=animals,)
plt.xlabel("Phylum")
plt.xticks(rotation=90)
plt.ylabel("Genome GC%")
ax.set_yscale('log')
plt.show()
```

```python
fig = plt.figure(figsize=(100, 30))
ax = plt.subplot(1, 3, 1)
sb.boxplot(x="Family", y="Genome.GC", data=animals,)
plt.xlabel("Family")
plt.xticks(rotation=90)
plt.ylabel("Genome GC%")
ax.set_yscale('log')
plt.show()
```
