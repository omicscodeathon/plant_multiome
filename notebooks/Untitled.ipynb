```python
from Bio import Entrez
import pandas as pd
import os
```

```python
def get_tax_id(species):
    species = species.replace(" ", "+").strip()
    search = Entrez.esearch(term = species, db = "taxonomy", retmode = "xml")
    record = Entrez.read(search)
    return record['IdList'][0]
#####################################################
def get_tax_data(taxids):
    dd = {'Superkingdom':[],
         'Phylum':[],
         'Class':[],
         'Order':[],
         'Family':[],
         'Genus':[]}
    for i in taxids:
        search = Entrez.efetch(id = i, db = "taxonomy", retmode = "xml")
        taxdata = Entrez.read(search)
        for taxa in taxdata[0]['LineageEx']:
            if taxa['Rank'] == 'superkingdom':
                dd['Superkingdom'].append(taxa['ScientificName'])
            elif taxa['Rank'] == 'phylum':
                dd['Phylum'].append(taxa['ScientificName'])
            elif taxa['Rank'] == 'class':
                dd['Class'].append(taxa['ScientificName'])
            elif taxa['Rank'] == 'order':
                dd['Order'].append(taxa['ScientificName'])
            elif taxa['Rank'] == 'family':
                dd['Family'].append(taxa['ScientificName'])
            elif taxa['Rank'] == 'genus':
                dd['Genus'].append(taxa['ScientificName'])
    
    return dd
```
